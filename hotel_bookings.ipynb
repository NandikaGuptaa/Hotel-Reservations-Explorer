```python
# combining "undefined" and "SC" meals as one due to their same meaning
data['meal'].replace('Undefined' , 'SC' , inplace = True)
```

```python
# checking invalid entries w.r.t adults, children, babies count
invalid_data = data[(data['adults'] == 0) & (data['children'] == 0) & (data['babies'] == 0)]
invalid_data[['adults' , 'children' , 'babies']]
```

```python
delete_data = (data['adults'] == 0) & (data['children'] == 0) & (data['babies'] == 0)
delete_data
```

```python
# converting false values into true values for 'delete_data' and storing it as new_data
new_data = data[~delete_data]
new_data
```

```python
# hotel reservations count
sns.countplot(x = 'hotel' , data = new_data)
plt.title('Total Hotel Reservations')
plt.show()
```

```python
# getting data of guests hotel wise who didnot cancel their reservations
resort_hotel = new_data[(new_data['hotel'] == 'Resort Hotel') & (new_data['is_canceled'] == 0)]
city_hotel = new_data[(new_data['hotel'] == 'City Hotel') & (new_data['is_canceled'] == 0)]
```

```python
# getting the combined guests data of both hotels who didnot cancel their reservations
booked_data = new_data[new_data['is_canceled'] == 0]
```

```python
# Prices guest pay for a room per night
sns.barplot(x = 'reserved_room_type' , y = 'adr' , hue = 'hotel' , data = booked_data)
plt.title('Prices of Room Types per night')
plt.xlabel('Room Type')
plt.ylabel('Price')
plt.show()
```

```python
# Prices guest pay for a room per night
sns.boxplot(x = 'reserved_room_type' , y = 'adr' , hue = 'hotel' , data = booked_data)
plt.title('Prices of Room Types per night')
plt.xlabel('Room Type')
plt.ylabel('Price')
plt.show()
```

```python
# prices variation over the year for 'resort hotel' 
resort_hotel_price = resort_hotel.groupby(['arrival_date_month'])['adr'].mean().reset_index()
month_to_num = { 'January' : 1 , 'February' : 2 , 'March' : 3 , 'April' : 4 , 'May' : 5 , 'June' : 6 , 'July' : 7 , 'August' : 8 , 'September' : 9 , 'October' : 10 , 'November' : 11 , 'December' : 12}
resort_hotel_price['arrival_date_month'] = resort_hotel_price['arrival_date_month'].map(month_to_num) 
resort_hotel_price
```

```python
resort_hotel_price.sort_values('arrival_date_month' , ascending = True , inplace = True)
```

```python
resort_hotel_price
```

```python
sns.barplot(x = 'arrival_date_month' , y = 'adr' , data = resort_hotel_price)
plt.title('ADR Per Month-Resort Hotel')
plt.xlabel('Month')
plt.ylabel('ADR')
plt.show
```

```python
# prices variation over the year for 'city hotel'
city_hotel_price = city_hotel.groupby(['arrival_date_month'])['adr'].mean().reset_index()
month_to_num = { 'January' : 1 , 'February' : 2 , 'March' : 3 , 'April' : 4 , 'May' : 5 , 'June' : 6 , 'July' : 7 , 'August' : 8 , 'September' : 9 , 'October' : 10 , 'November' : 11 , 'December' : 12}
city_hotel_price['arrival_date_month'] = city_hotel_price['arrival_date_month'].map(month_to_num) 
city_hotel_price
```

```python
city_hotel_price.sort_values('arrival_date_month' , ascending = True , inplace = True)
```

```python
city_hotel_price
```

```python
sns.barplot(x = 'arrival_date_month' , y = 'adr' , data = city_hotel_price)
plt.title('ADR Per Month-City Hotel')
plt.xlabel('Month')
plt.ylabel('ADR')
plt.show
```

```python
final_merged_hotels_adr = resort_hotel_price.merge(city_hotel_price , on = 'arrival_date_month')
final_merged_hotels_adr.columns = ['month' , 'ADR of Resort Hotel' , 'ADR of City Hotel']
final_merged_hotels_adr
```

```python
# guests per month in resort hotel to find which month was the busiest
busy_resort = resort_hotel['arrival_date_month'].value_counts().reset_index()
busy_resort.columns = ['month' , 'number of guests']
month_to_num = { 'January' : 1 , 'February' : 2 , 'March' : 3 , 'April' : 4 , 'May' : 5 , 'June' : 6 , 'July' : 7 , 'August' : 8 , 'September' : 9 , 'October' : 10 , 'November' : 11 , 'December' : 12}
busy_resort['month'] = busy_resort['month'].map(month_to_num) 
busy_resort
```

```python
busy_resort.sort_values('month' , ascending = True , inplace = True)
```

```python
busy_resort
```

```python
sns.barplot(x = 'month' , y = 'number of guests' , data = busy_resort)
plt.title('Guests Per Month in Resort Hotel')
plt.show()
```

```python
# guests per month in city hotel to find which month was the busiest
busy_city = city_hotel['arrival_date_month'].value_counts().reset_index()
busy_city.columns = ['month' , 'number of guests']
month_to_num = { 'January' : 1 , 'February' : 2 , 'March' : 3 , 'April' : 4 , 'May' : 5 , 'June' : 6 , 'July' : 7 , 'August' : 8 , 'September' : 9 , 'October' : 10 , 'November' : 11 , 'December' : 12}
busy_city['month'] = busy_city['month'].map(month_to_num) 
busy_city
```

```python
busy_city.sort_values('month' , ascending = True , inplace = True)
```

```python
busy_city
```

```python
sns.barplot(x = 'month' , y = 'number of guests' , data = busy_city)
plt.title('Guests Per Month in City Hotel')
plt.show()
```

```python
final_merged_busy_hotel = busy_resort.merge(busy_city , on = 'month')
final_merged_busy_hotel.columns = ['month' , 'no. of guests in resort hotel' , 'no. of guests in city hotel']
final_merged_busy_hotel
```
